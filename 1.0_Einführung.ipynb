# Einführung

## Motivation der Arbeit

Das innovative Konzept der Einfachheit und Übersichtlichkeit macht Python seit Jahren zu eines der beliebtesten Einsteigerprogrammiersprachen und wird zunehmend an Hochschulen und Universitäten angeboten. Python ist kostenlos nutzbar und bietet eine umfangreiche Standardbibliothek, die mit einer Vielzahl an Zusatzmodulen erweitert werden kann. Somit präsentiert Python für fast jedes erdenkliche wissenschaftliche Problem eine zugeschnittene Sammlung von Funktionen an und macht das Programmieren mit Python vor allem in den Bereichen der Datenverarbeitung, des Web-Developments und der Ingenieurswissenschaften zu einer spannenden Innovation. Mit wenig Aufwand können effiziente grafische Benutzeroberflächen erstellt oder Ergebnisse umfangreicher Berechnungen übersichtlich dargestellt werden (vgl. Steyer 2017: 3).

Im Bereich des kommerziellen Bauwesens wird Python in Deutschland trotz alledem erst sehr wenig angewandt. Besonders effizient lässt sich das Programmieren bei sich wiederholenden Prozessen einsetzen. Das Automatisieren einfacher Berechnungen oder kognitiver Routinearbeiten ersparen langfristig gesehen eine Menge Zeit und minimieren dabei die Fehleranfälligkeit. 

Die Bearbeitung von immer größeren Datenmengen im Zeitalter der Digitalisierung stellt eine Herausforderung dar, die auch im Bauwesen zu bewerkstelligen gilt. Die rasante technologische Entwicklung in allen beruflichen Bereichen wird dazu führen, dass es ebenso für die Ingenieure wichtig werden wird, Kenntnisse im Programmieren zu erlernen. Seit der Einführung von computergestützten statischen Berechnungsverfahren entwickelt sich der Anspruch im Bauwesen dahingehend, dass Bauwerke zunehmend komplexer und anspruchsvoller werden. Die hohen Anforderungen werden aus den Folgen des Klimawandels, wie die daraus resultierende Ressourcenknappheit befeuert, sodass ökonomisches und effizientes Bauen unter Einbeziehung ökologischer Aspekte wichtiger ist denn je.

Im Zuge dessen erfährt auch der Holzbau aufgrund seiner Nachhaltigkeit wieder zunehmend an Popularität und an Vertrauen in der Gesellschaft. Als einzig erneuerbarer Bauwerkstoff wird mit dem Einsatz von Holz eine nachhaltige Entscheidung getroffen, die viele Vorteile mit sich bringt. Neben einer sehr guten ökologischen Bilanz besitzt Konstruktionsvollholz zudem hervorragende mechanische Eigenschaften und schafft ein gesundes Raumklima. Eines der entscheidenden Eigenschaften des Materials ist vor allem die Bindung des klimaschädlichen Kohlenstoffs und dessen Umwandlung in Sauerstoff. Dadurch leistet Holz einen wichtigen Beitrag zum Klimawandel (vgl. Neuhaus 2017: 1). 

Aus diesem Grund werden im Rahmen dieser Arbeit das Potenzial von Python mit Hilfe verschiedener Anwendungsmöglichkeiten praxisnah erläutert. Das Ziel besteht darin, Stärken und Schwächen von Python anhand dieser Anwendungsbeispiele aufzuzeigen und dabei eine Schritt-für-Schritt Anleitung zu erstellen. Die Beispiele sind so gewählt, dass sie möglichst simpel in die alltägliche Arbeitsweise eines Statikers integriert und zu eigenen Anwendungen und Skripten weiterentwickelt werden können.

## Herangehensweise

Das Ziel dieser Arbeit ist darin definiert, dass es grundsätzlich als Einsteigerkurs für die Programmiersprache Python konzipiert ist. Anhand von Anwendungsbeispielen werden die Grundlagen des Programmierens aufgezeigt und verschiedene Algorithmen sukzessive aufgebaut, sodass die Arbeit zusammenhängend gelesen werden sollte.

Für das Programmieren von eigenen Anwendungen mit Python sind zu Beginn nur eine Handvoll an Grundlagen notwendig, die immer wieder zum Einsatz kommen. Es gibt ein unzähliges Angebot an verschiedenen kostenlosen Seminaren und Tutorien im Internet, die auf Deutsch und Englisch angeboten werden, die die theoretischen Grundlagen ausführlich und umfangreich erklären. Daher wird im Rahmen dieser Arbeit ein praktischer Ansatz gewählt, wobei das Programmieren mit bekannten ingenieurtechnischen Problemstellungen aufgeschlüsselt wird. Anhand dessen wird der Leser dazu angehalten, trotz weniger Vorkenntnisse, die angeführten Beispiele für eigene Zwecke anzuwenden. Erst durch die praktische Anwendung sind schnelle und nachhaltige Erfolge garantiert.

### Integration von Python in Excel

Als Ingenieur ist es stets von großer Bedeutung, die Berechnungen inhaltlich zu verstehen und Ergebnisse anhand des eigenen Fachwissens zu hinterfragen. Trotz der Anwendung diverser Programme und Softwares sollte das Verständnis und das Vertrauen der Berechnungen niemals geringer ausfallen.

Das Arbeiten mit dem Tabellenkalkulationsprogramm Microsoft Excel bietet aus diesen Gründen enorme Vorteile als Ingenieur, um leistungsstark und effizient Dateien zu erstellen und zu bearbeiten. Es können intuitive Berechnungstabellen erzeugt werden, die flexibel angepasst und von anderen Nutzern leicht nachvollzogen werden können. Vor allem in der Buchhaltung, der Finanzwelt und im Ingenieurswesen ist Excel nicht mehr weg zu denken und genießt ein hohes Maß an Vertrauen bei den Nutzern. 

Bei komplexeren und größeren Datenmengen kann es jedoch dazu führen, dass sich die Ausführung von Excel verlangsamt und das Arbeiten dadurch stark beeinträchtigt wird. Eines der Hauptursachen stellen eine hohe Anzahl an Formeln und Zellenreferenzen dar, die in großen Tabellen dazu führt, dass die Geschwindigkeit der Datenverarbeitung dadurch abnimmt. 

Aus diesem Grund soll im ersten Anwendungsbeispiel aufgezeigt werden, wie durch das Verknüpfen von Excel mit Python ein Stützenbemessungsprogramm entsteht, welches die aufgeführten Aspekte miteinander vereint. Im Sinne der Nachvollziehbarkeit sollen Zwischenergebnisse angezeigt werden und es soll hinzu ermöglicht werden, dass durch das Nutzen von Eingabeparametern eine Interaktion mit dem Nutzer erzielt wird. Die Berechnung selbst übernimmt das Python-Skript, welches innerhalb der Excel-Oberfläche mit einem Schalter ausgeführt wird. Das Ergebnis wird anschließend an Excel übermittelt, sodass keine komplizierten Formeln benötigt werden. 

Das Stützenbemessungsprogramm soll einen gesamten Arbeitsablauf von der Bemessung, Optimierung, bis hin zur Dokumentation abbilden.

### Entwicklung einer Webapplikation mit Streamlit

In diesem Anwendungsbeispiel soll ein Programm entwickelt werden, welches die Tragfähigkeiten von Vollgewindeschrauben des Herstellers Würth berechnet. Die Berechnungen der Tragfähigkeiten eines Verbindungsmittels hinsichtlich einer Axial- und Scherbeanspruchung ist in der DIN EN 1995-1-1 Absatz 8 geregelt. Durch die hohe Anzahl der Einflussparameter und herstellerspezifischen Materialkennwerte erweist sich die Bemessung als vergleichsweise komplex und kompliziert. Aus diesem Grund gibt es herstellereigene Programme, die verschiedene Holzverbindungen nachweisen. Als Alternative werden Bemessungstafeln angeboten, anhand dessen die Axial- und die Schertragfähigkeiten der Schrauben unter Berücksichtigung gewisser Randbedingungen, wie Seitenholzdicke oder Holzgüte abgelesen werden können. 

Die Nutzung der herstellereigenen Programme und Tabellen ist vor allem bei häufiger Nutzung zeitintensiv und stellt eine Arbeitsweise dar, die verbesserungsfähig ist. Mit der Python-Bibliothek Streamlit soll daher eine Webapplikation für die Bemessung von Vollgewindeschrauben entwickelt werden. 

Bei *Streamlit* handelt es sich um eine Bibliothek eines Drittanbieters, der sich mit interaktiven webbasierten Applikationen beschäftigt. Mit wenig Aufwand können benutzerfreundliche Oberflächen erstellt werden, die anschließend als Programm exportiert und öffentlich geteilt werden können. Streamlit soll daher genutzt werden, um ein Bemessungsprogramm zu schreiben, welche die Schwächen der existierenden Tools aufgreift und eine simple Möglichkeit darbietet Vollgewindeschrauben zu bemessen.
