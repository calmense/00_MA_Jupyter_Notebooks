# Vorwort

Die vorliegende Abschlussarbeit behandelt zwei Themenbereiche des Holzingenieurbaus, die ich als junger Ingenieur in der Berufspraxis behandelt habe.

Zum einen bestand meine Aufgabe im Rahmen einer Planung eines Bürogebäudes darin, etwa eintausend Stützen zu bemessen. Nach einem Erkunden für die gängige Vorgehensweise bei dieser Aufgabe, war ich erstaunt, dass dies üblicherweise mit einer aufwendigen Typisierung bewerkstelligt wird. Die Typisierung aller Stützen des Gebäudes ergaben ca. 60 verschiedene Stützentypen, die je nach Lage, Abmessungen, Länge und Einwirkungen gruppiert wurden. Berechnet wurden sie mit Einzelnachweisen mithilfe des Programms Frilo (Friedrich + Lochner). Zusammen mit der Dokumentation für die abzugebende prüfbare Statik beendete ich diese Aufgabe in annähernd drei Monaten.

Im weiteren Verlauf des Projektes war es anschließend meine Aufgabe, mich den Bauteilanschlüssen zu widmen. Das Ziel dabei war es, im Sinne einer Rückbaubarkeit möglichst einfache Stahlverbindungen mit Vollgewindeschrauben zu konzipieren. Die Bemessung der Holzschrauben erfolgte anhand firmenspezifischer Produktblätter, die teilweise aus über hundert Seiten bestanden.

Wieder sah ich mich also in einer Situation, in der ich dieselben Arbeitsschritte teils hundertfach wiederholen musste. Sowohl bei der Stützenbemessung als auch bei der Schraubenbemessung erkannte ich daher ein großes Potenzial für die Automatisierung des Arbeitsablaufes. Dies würde nicht nur in einem geringeren Arbeitsaufwand resultieren, sondern würde ebenfalls eine viel genauere Bemessung, eine deutlich höhere Zeiteffizienz und dadurch hohe Kosteneinsparungen ergeben.

Aus diesen praktischen Erfahrungen heraus ergab sich mir die Motivation, die Anwendbarkeit von Python anhand dieser Anwendungsbeispiele zu evaluieren und aufzuzeigen, wie die Produktivität der Statik mit der Unterstützung Pythons gesteigert werden kann. Die Abschlussarbeit begann ich mit sehr wenigen Vorkenntnissen im Programmieren, sodass ich mir sämtliche Kenntnisse während der Bearbeitung angeeignet habe.  Besonders wichtig war mir eine möglichst zugängliche und interaktive Aufbereitung der Materie, um der weitverbreiteten Befangenheit gegenüber dem Programmieren zu entgegnen.
