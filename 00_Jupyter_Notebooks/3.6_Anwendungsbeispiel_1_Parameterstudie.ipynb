### Theorie II. Ordnung: Gegenüberstellung mit Näherungslösung

Für die Python-gestützte Schnittgrößenermittlung nach der Theorie II. Ordnung wurde der Zuwachs der Verformung durch die aus der Festigkeitslehre bekannten Formeln der Durchbiegung berechnet. Zur Kontrolle soll dieses Verfahren mit einer Näherungslösung gegenübergestellt werden, die nach Neuhaus wie folgt für eine Pendelstütze angeführt wird (2017: 823):

<center> $\alpha = 1/(1-\delta)$ <br>
<center> $\delta = N/(N_{crit})$ <br>
<center> $N_{crit} = (\pi^2*E*I)/L^2$ <br>
    <center> $M_{II} = M_{I} + N*e*\alpha$<br> </center>
    
Der $\alpha$-Vergrößerungsfaktor bestimmt den Momentenzuwachs infolge der unvermeidbaren Vorvervormung des Stabes e.
Die Gegenüberstellung der Ergebnisse für das Moment nach Theorie II. Ordnung zeigen eine hohe Genauigkeit für Schlankheitswerte unter 50. Für Schlankheiten über 50 steigt das Moment für das p-Delta Verfahren stärker an (s. Grafik).

### Gegenüberstellung der Bemessungsverfahren

Abschließend lässt sich sagen, dass im Bereich von mittleren Schlankheiten (20> $\lambda$ < 100) alle drei Bemessungsverfahren vergleichbare Ergebnisse erzielen, wobei die Tragfähigkeiten eines Stabes, welcher nach der Theorie II. Ordnung bemessen wurde 10% bis 20% wirtschaftlicher ausfällt. Der Grund für den Unterschied ist nach Neuhaus auf die Ermittlung des Knickbeiwertes zurückzuführen. Hierbei wird nach der DIN EN 1995-1-1 der bezogene Schlankheitsgrad mit einer charakteristischen Druckfestigkeit berechnet, sodass der Abminderungsbeiwert $k_c$ und resultierend die Tragfähigkeit kleiner ausfällt (2017: 832). 

Bei Schlankheitswerten größer als 100 zeigt der Vergleich mit beiden Verfahren, dass bei sehr schlanken Bauteilen die Schnittgrößenermittlung nach der Theorie II. Ordnung von großer Bedeutung ist.
